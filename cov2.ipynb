```python
from matplotlib import pyplot as plt
import numpy as np
from ipywidgets import interact
import matplotlib.patches as mpatches
from IPython.display import clear_output
```

```python
N = np.random.randn(100, 2)
N[:,0] *= 2.0
    
def vis_pca(theta):
    # rotate
    A = np.array([[np.cos(theta), -np.sin(theta)],
                  [np.sin(theta), np.cos(theta)]])
    
    Nrotated = N.dot(A)
    
    cov = Nrotated.T.dot(Nrotated)/Nrotated.shape[0]
    val, vec = np.linalg.eig(cov)
    
    plt.figure(figsize=(6, 6))
    plt.axis([-8, 8, -8, 8])
    plt.gca().set_aspect('equal')
    fig = plt.scatter(Nrotated[:,0], Nrotated[:,1])
    
    plt.plot([0, vec[0, 0]*val[0]], [0, vec[1, 0]*val[0]], c='r', linewidth=3)
    plt.plot([0, vec[0, 1]*val[1]], [0, vec[1, 1]*val[1]], c='g', linewidth=3)
    
interact(vis_pca, theta=(0, np.pi*2.0, .1))
```

```python
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)

imgs = mnist.train.images.copy()

# zero mean
for j in range(imgs.shape[1]):
    imgs[:,j] -= np.sum(imgs[:,j])/imgs.shape[0]
    
# associate each with a color
cols = ['red', 'green', 'blue', 'orange', 'black', 'gray', 'purple', 'brown', 'pink', 'tan']
colors = [cols[np.argmax(mnist.train.labels[i])] for i in range(mnist.train.labels.shape[0])]
```

```python
def vis_digit(i):
    plt.figure(figsize=(6, 6))
    plt.imshow(imgs[i].reshape(28, 28), interpolation='nearest', cmap=plt.cm.gray)
    
    col = np.argmax(mnist.train.labels[i])
    recs = [mpatches.Rectangle((0,0),1,1,fc=cols[col])]
    plt.legend(recs,[col],loc=4)
    #print(colors[i])

interact(vis_digit, i=(0, imgs.shape[0], 1))
```

```python
# neural net with 2 hidden units, softmax classifier, this is cheating because we know the classes

learning_rate = 0.001

hdim = 10
W1 = np.random.randn(imgs.shape[1] + 1, hdim) * .1
W2 = np.random.randn(hdim + 1, 10) * .1

for i in range(50000):
    xi = np.random.randint(imgs.shape[0])
    t = mnist.train.labels[xi]
    
    x = imgs[xi].reshape(1, 784)
    x = np.concatenate((x, [[1]]), axis=1)
    z1 = x.dot(W1)
    
    h1 = 1.0 / (1.0 + np.exp(-z1))
    h1 = np.concatenate((h1, [[1]]), axis=1) # 1 x 33
    
    z2 = h1.dot(W2)
    
    z2_sum = np.sum(np.exp(z2))
    h2 = np.exp(z2) / z2_sum # 1 x 10
    
    dz2 = h2 - t
    gradW2 = np.multiply(h1.T, dz2)
    dh1 = W2.dot(dz2.T).T
    dz1 = np.multiply(h1*(1-h1), dh1)
    gradW1 = np.multiply(x.T, dz1[:, :-1])
    
    W1 -= gradW1 * learning_rate
    W2 -= gradW2 * learning_rate
    
    W1 *= 0.99998
    W2 *= 0.99998
    
    if i % 1000 == 0:
        clear_output(wait=True)
        #print('prediction: {} true: {}'.format(np.argmax(h2), np.argmax(t)))
        plt.figure(figsize=(10, 10))
        vis_nn_pca(dim0=0, dim1=1)
        plt.show()
```

```python
def test(i):
    xi = np.random.randint(imgs.shape[0])
    t = mnist.train.labels[xi]
    x = imgs[xi].reshape(1, 784)
    
    xb = np.concatenate((x, [[1]]), axis=1)
    z1 = xb.dot(W1)
    
    h1 = 1.0 / (1.0 + np.exp(-z1))
    h1 = np.concatenate((h1, [[1]]), axis=1) # 1 x 33
    
    z2 = h1.dot(W2)
    
    z2_sum = np.sum(np.exp(z2))
    h2 = np.exp(z2) / z2_sum # 1 x 10
    
    fig = plt.imshow(x.reshape(28, 28), cmap=plt.cm.gray)
    print('prediction: {}'.format(np.argmax(h2)))
    
interact(test, i=(1, imgs.shape[0], 1))
```

```python
def vis_nn_pca(dim0, dim1):
    points = 1000

    l = mnist.train.labels[0:points]
    colors = [cols[np.argmax(l[i])] for i in range(points)]

    x = imgs[0:points]
    x = np.concatenate((x, np.ones([points, 1])), axis=1) # points x xdim+1
    z1 = x.dot(W1)

    #h1 = 1.0 / (1.0 + np.exp(-z1))
    h1 = z1
    
    plt.gca().set_aspect('equal')
    fig = plt.scatter(h1[:,dim0], h1[:,dim1], color=colors)
    
interact(vis_nn_pca, dim0=(0, hdim, 1), dim1=(0, hdim, 1))
```
